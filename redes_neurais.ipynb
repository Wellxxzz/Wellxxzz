```python
import pandas as pd
```

```python
df = pd.read_excel("/content/churn.xlsx")
df.head(10)
```

```python
df = df.drop(["customerID"], axis=1)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.dtypes
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
cats_vars=["PhoneService","Contract","PaperlessBilling","PaymentMethod"]
num_vars=["tenure","MonthlyCharges","TotalCharges"]
modeling_vars=cats_vars+num_vars
```

```python
x = df.filter(modeling_vars).copy
y= df["Churn"].copy
```

```python
y= pd.DataFrame()
```

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()

le.fit(df["Churn"])
y["Churn"]=le.transform(df["Churn"])
```

```python
y
```

```python
x= pd.get_dummies(df[cats_vars])
```

```python
x
```

```python
df["Churn"].value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100, 50, 20), activation= "logistic")
clf.fit(x_train_scaled, y_train)
```

```python
y_pred = clf.predict(x_test_scaled)
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100} %")
print(f"F1 Score: {f1_score(y_test, y_pred)*100} %")
```
